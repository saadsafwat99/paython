```python
import networkx as nx
G=nx.Graph()
G.add_nodes_from([1,2,3,4,5 ,6])
G.add_edges_from([(1,2),(1,3),(1,4),(1,6),(2,3),(2,4),(3,6)])

nx.draw(G,with_labels=True,node_color='red',node_size=500,font_color='white',font_size=25)
a = nx.adjacency_matrix(G).todense()
print("Adjacency matrix is : ")
print(a)
```

```python
import matplotlib.pyplot as plt
import networkx as nx    
from collections import Counter

degree_sequence = [G.degree(n) for n in G.nodes]
degree_counts = Counter(degree_sequence)
min_d, max_ = min(degree_counts.keys()), max(degree_counts.keys())
plt.xlabel("Degree", fontsize=20)
plt.ylabel("Number of Nodes", fontsize=20)
plot_x = list(range(min_d, max_d + 1))
plot_y = [degree_counts.get(x, 0) for x in plot_x]
plt.bar(plot_x, plot_y)

def plot_degree_dist(G):
    
    degrees = G.degree()
    degrees = dict(degrees)
    values = sorted(set(degrees.values()))
    print(values)
    histo = [list(degrees.values()).count(x) for x in values]
    p = [x / G.order() for x in histo]
    print(len(p))
    
    plt.figure()
    plt.bar(values, p)
    plt.xlabel("k",fontsize=20)
    plt.ylabel("p(k)", fontsize=20)
    plt.title("Degree Distribution", fontsize=20)
    
    plt.show()
    plt.figure()
    plt.grid(False)
    plt.loglog(values, p, "bo")
    plt.xlabel("k", fontsize=20)
    plt.ylabel("log p(k)", fontsize=20)
    plt.title("log Degree Distribution")
    plt.show()
    plt.show() 
```

```python
plot_degree_dist(G)
```

```python
D=nx.DiGraph()
D.add_nodes_from([1,2,3,4,5 ,6])
D.add_edges_from([(2,3),(3,2),(2,4),(6,3),(1,2),(6,1),(4,1),(3,1)])
nx.draw(D,with_labels=True,node_color='red',node_size=500,font_color='white',font_size=25)
a = nx.adjacency_matrix(D).todense()
print("Adjacency matrix is : ")
print(a)
```

```python
plot_degree_dist(D)
```
